```python
df = urllib.request.urlopen('https://simple.wikipedia.org/wiki/Cat').read()
```

```python
df
```

```python
df_paragraph = bs.BeautifulSoup(df,'html').find_all('p')
```

```python
df_paragraph
```

```python
##craeting the corpus
cat_text=''
```

```python
#creating lower text corpus of df_paragraph
for p in df_paragraph:
    cat_text = cat_text+ p.text.lower()
    
print(cat_text)
```

```python
## to clean up the text and bracket number and empty spaces
import re
##re.sub('[^a-zA-Z]',' ',cat_text)
```

```python
cat_text = re.sub(r'\s+' ,' ',re.sub(r'\[[0-9]*\]' ,' ',cat_text))
```

```python
cat_text
```

```python
## create a list of sentences based on corpus
import nltk
nltk.download() ## first time use only
nltk.download('punkt') ## first time use only
cat_sentences = nltk.sent_tokenize(cat_text)
cat_sentences
```

```python
## cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def chatbot_answer(user_query):
    cat_sentences.append(user_query)
    cv =TfidfVectorizer()
    vectors = cv.fit_transform(cat_sentences)
    vectors_value = cosine_similarity(vectors[-1],vectors)
    answer = cat_sentences[vectors_value.argsort()[0][-2]]
    
    input_check = vectors_value.flatten()
    input_check.sort()
    
    if input_check[-2] ==0:
        return "try again"
    else:
        return answer
    
```

```python
print("hello,i am chabot.what is your question?")
while(True):
    query=input().lower()
    if query not in['bye','good bye','take care']:
        print("chatbot :",end="")
        print(chatbot_answer(query))
        cat_sentences.remove(query)
        
    else:
        print("see you again")
        break
```
